```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def parse_asf(filename):
    skeleton = {}
    with open(filename, 'r') as f:
        lines = f.readlines()
        
    hierarchy = {}
    in_hierarchy_section = False
    bone = None
    for line in lines:
        line = line.strip()
        
        if line.startswith(':hierarchy'):
            in_hierarchy_section = True
                    
        elif line.startswith('begin') and not in_hierarchy_section:
            bone = {}
        elif line.startswith('end') and not in_hierarchy_section:
            if 'name' in bone:
                skeleton[bone['name']] = bone
            bone = None
        elif bone is not None:
            parts = line.split()
            if parts[0] == 'id':
                bone['id'] = int(parts[1])
            elif parts[0] == 'name':
                bone['name'] = parts[1]
            elif parts[0] == 'direction':
                bone['direction'] = np.array([float(v) for v in parts[1:]])
            elif parts[0] == 'length':
                bone['length'] = float(parts[1])
            elif parts[0] == 'axis':
                bone['axis'] = [float(v) for v in parts[1:4]]
            elif parts[0] == 'axis':
                bone['axis'] = [float(v) for v in parts[1:4]]
                
        elif in_hierarchy_section and not line.startswith(('begin', 'end')):
            segments = line.split()
            parent = segments[0]
            children = segments[1:]

            if parent not in hierarchy:
                hierarchy[parent] = []
                
            hierarchy[parent].extend(children)
            
    return skeleton, hierarchy

def build_skeleton(skeleton, hierarchy):
    joints = {}
    skeleton['root'] = {}
    skeleton['root']['position'] = np.array([0, 0, 0])
    
    for parent, childs in hierarchy.items():
        parent_position = skeleton[parent]['position']
        for child in childs:
            direction = skeleton[child]['direction']
            length = skeleton[child]['length']
            direction_norm = direction / np.linalg.norm(direction)
            child_position = parent_position + direction_norm * length
            joints[child] = (parent_position, child_position)
            skeleton[child]['position'] = child_position
    
    return joints

def draw_skeleton(joints):
    fig = plt.figure(figsize=(16, 10))
    ax = fig.add_subplot(111, projection='3d')

    for joint, (start, end) in joints.items():
        xs, ys, zs = zip(start, end)
        ax.plot(xs, ys, zs, marker='o', label=joint)
    
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    ax.set_title('3D Skeleton Visualization')
    plt.legend()
    plt.show()
```

```python
skeleton, hierarchy = parse_asf(sample_asf_path)
joints = build_skeleton(skeleton, hierarchy)
draw_skeleton(joints)
```

```python
import plotly.graph_objects as go

def draw_skeleton_with_plotly(joints):
    fig = go.Figure()

    for joint, (start, end) in joints.items():
        xs, ys, zs = zip(start, end)
        
        fig.add_trace(go.Scatter3d(
            x=xs, 
            y=zs, 
            z=ys, 
            mode='lines+markers',
            marker=dict(size=5),
            line=dict(width=4),
            name=joint
        ))

    fig.update_layout(
        scene=dict(
            xaxis_title='X axis',
            yaxis_title='Y axis',
            zaxis_title='Z axis',
            xaxis=dict(range=[-5, 5]),
            yaxis=dict(range=[-5, 5]),
            zaxis=dict(range=[-5, 5]),
        ),
        title='3D Skeleton Visualization',
        width=800,
        height=600,
    )

    fig.show()
```

```python
draw_skeleton_with_plotly(joints)
```

```python
def parse_amc_file(amc_file_path):

    joint_data = {}
    with open(amc_file_path, 'r') as file:
        lines = file.readlines()
    
    current_frame = 0
    for line in lines:
        line = line.strip()
        if not line or line.startswith(":"):
            continue

        parts = line.split()
        if parts[0].isdigit():
            current_frame = int(parts[0])
            joint_data[current_frame] = {}
            continue
        
        joint_name = parts[0]
        joint_values = list(map(float, parts[1:]))
        joint_data[current_frame][joint_name] = joint_values
    
    return joint_data
```

```python
frames = parse_amc_file(sample_amc_path)
frames[10]
```

```python
def prepare_frame_view(skeleton, hierarchy, frames, frame_idx):
    new_joints = {}
    skeleton['root']['position'] = frames[frame_idx]['root'][:3]
    root_direction = np.array(frames[frame_idx]['root'][3:])
    
    for parent, childs in hierarchy.items():
        parent_position = skeleton[parent]['position']
        for child in childs:
            direction = root_direction + skeleton[child]['direction'] + np.array(frames[frame_idx].get(child, [0, 0, 0]))
            length = skeleton[child]['length']
            direction_norm = direction / np.linalg.norm(direction)
            child_position = parent_position + direction_norm * length
            new_joints[child] = (parent_position, child_position)
            skeleton[child]['position'] = child_position
    
    return new_joints
```

```python
joints_move = prepare_frame_view(skeleton, hierarchy, frames, 10)
joints_move
```

```python
draw_skeleton_with_plotly(joints_move)
```
